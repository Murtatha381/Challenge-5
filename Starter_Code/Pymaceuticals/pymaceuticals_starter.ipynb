```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy import stats
from scipy.stats import linregress
import numpy as np
from scipy.stats import iqr


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")


# Display the data table for preview
merged_df.head()
```

```python
# Checking the number of mice.
merged_df['Mouse ID'].count()
#merged_df['Mouse ID'].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
Mouse_df = merged_df[merged_df.duplicated()]
Mouse_df
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
Mouses_df = merged_df.drop_duplicates()
Mouses_df.head()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
Mouses_df.head(15)
```

```python
# Checking the number of mice in the clean DataFrame.
Mouses_df['Mouse ID'].count()
#Mouses_df['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_df = Mouses_df[["Drug Regimen","Tumor Volume (mm3)"]]
mean = summary_df.groupby(["Drug Regimen"]).mean().head(10)
median = summary_df.groupby(["Drug Regimen"]).median().head(10)
var = summary_df.groupby(["Drug Regimen"]).var().head(10)
stan_dev = summary_df.groupby(["Drug Regimen"]).std().head(10)
standard_err = summary_df.groupby(["Drug Regimen"]).sem().head(10)
index = summary_df["Drug Regimen"].unique()
sum_stats = summary_df.groupby(["Drug Regimen"]).describe()
sum_stats['median']= median
sum_stats['average']= mean
sum_stats['variance']= var
sum_stats['STD']= stan_dev
sum_stats['SEM']= standard_err
summary_stats= sum_stats[['median','average','variance','STD','SEM']]
summary_stats

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
#Summ_df = pd.DataFrame({
                        # "Drug": index,
                        # "mean": mean,
                        # "median": median,
                        # "variance": var,
                        #"standard deviation": stan_dev,
                        # "SEM": standard_err})
                      
```
